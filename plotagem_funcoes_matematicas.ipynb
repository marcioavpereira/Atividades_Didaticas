#Problema 1
Criar uma plotagem para $e^{-x} sen(4x)$ e $e^{-x} cos(4x)$ de 0 a 5.
Mudar a cor e largura de linhas para vermelho e azul, 4.0 e 2.0, respectivamente. Exibir legenda e títulos.

Importação das bibliotecas

```
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

Criação de valores de 0 a 5 em x, funções seno e cosseno em y e z, e plotagem.

```
x = np.linspace(0,5,100)
y = np.exp(-x) * np.sin(4*x)
z = np.exp(-x) * np.cos(4*x)
plt.figure(figsize=(15,8))
plt.plot(x,y, 'r--', label = '$e^{-x} sin(4x)$', linewidth=4.0)
plt.plot(x,z, label = '$e^{-x} cos(4x)$', linewidth=2.0)
plt.title('Gráfico para as funções seno e cosseno')
plt.xlabel('Valores de entrada x')
plt.ylabel('Valores de saída das funções y e z')
plt.legend();
plt.savefig('Funções Seno e Cosseno.png')
```
