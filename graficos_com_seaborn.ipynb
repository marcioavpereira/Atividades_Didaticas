# Análises com Variações Gráficas usado Seaborn

>Importação de bibliotecas gráficas

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

>Carregando o conjunto de dados Iris

```python
iris = sns.load_dataset('iris')
iris
```

>Plotando a dispersão de combinações das variáveis pétala e sépala, comprimento e largura.

```python
sns.PairGrid(iris, hue='species', palette='husl').map(plt.scatter).add_legend();
```

>Plotando a dispersão com histogramas na diagonal e outra paleta de cores.

```python
sns.PairGrid(iris, hue='species', palette='Set1').map_diag(plt.hist).map_offdiag(plt.scatter).add_legend();
```

>Plotando apenas histogramas customizados na diagonal.

```python
sns.PairGrid(iris).map_diag(plt.hist, facecolor='yellow', edgecolor='red', linewidth=3);
```

>Plotando uma variação

```python
sns.PairGrid(iris, hue='species', palette='coolwarm').\
                                                      map_diag(plt.hist).map_offdiag(plt.scatter, edgecolor='yellow').\
                                                      add_legend();
```

>Plotando uma divisão de tipos

```python
sns.PairGrid(iris).map_diag(plt.hist, facecolor='purple', edgecolor='yellow', linewidth=2).\
                   map_upper(plt.scatter).map_lower(sns.kdeplot);
```

>Mais uma variação de plotagem

```python
f, ax = plt.subplots(figsize=(15, 7))
sns.heatmap(iris.corr(), annot=True, cmap="RdYlGn");
```
