# Importação e Visualização de Múltiplas Imagens

```python
# importação de bibliotecas

from skimage.io import imread
import matplotlib.pyplot as plt
```

```python
# Obtenção e Visualização de uma Amostra de Imagem

imagem = imread('C:/Users/Karla/Documents/Marcio/Pintura/Desenho/Casal/Casal 1.jpg')
plt.imshow(imagem)
plt.show()
```

```python
# laços FOR para importar e exibir as imagens

endereco = 'C:/Users/Karla/Documents/Marcio/Pintura/Desenho/Casal/'
imagens =[]

for i in range(10):
    imagem = imread(endereco + 'Casal ' + str(i+1) + '.jpg')
    imagens.append(imagem)

figura = plt.figure(figsize=(40,30))

for i in range(10):
    figura.add_subplot(4,3, i+1)
    plt.imshow(imagens[i])
plt.show()
```
